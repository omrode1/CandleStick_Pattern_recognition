```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker = "AAPL"  # Example: Apple Inc.

# Get data from Yahoo Finance
data = yf.download(ticker, start="2020-01-01", end="2023-08-01", interval="1d")

# Display the data
print(data.head())
```
